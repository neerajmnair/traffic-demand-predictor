```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv(
    "../data/train.csv"
)

test_df = pd.read_csv(
    "../data/test.csv"
)
```

## Demand Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(
    train_df["demand"],
    bins=30,
    kde=True
)

plt.title("Traffic Demand Distribution")
plt.xlabel("Demand")
plt.ylabel("Frequency")

plt.show()
```

## Temperature vs Demand

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    train_df,
    x="Temperature",
    y="demand"
)

plt.title("Temperature vs Traffic Demand")

plt.show()
```

## RoadType vs Demand

```python
plt.figure(figsize=(8,5))

sns.boxplot( 
    train_df,
    x="RoadType",
    y="demand")
plt.title("Road Type vs Demand")
plt.xticks(rotation=45)

plt.show()
```

RoadType shows a strong relationship with traffic demand.
Highways have the highest median demand, followed by Streets, while Residential roads have the lowest demand. This suggests that RoadType is an important feature for predicting traffic demand.

## Weather vs Demand

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    train_df,
    x="Weather",
    y="demand"
)

plt.title("Weather vs Traffic Demand")
plt.xticks(rotation=45)

plt.show()
```

## Number of Lanes vs Demand

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    train_df,
    x="NumberofLanes",
    y="demand"
)

plt.title("Number of Lanes vs Demand")

plt.show()
```

Roads with 4 and 5 lanes exhibit substantially higher traffic demand than roads with 1–3 lanes. The number of lanes appears to be an important predictor of traffic demand and may be associated with road type.

```python
train_df["hour"] = pd.to_datetime(
    train_df["timestamp"],
    format="%H:%M"
).dt.hour
```

```python
plt.figure(figsize=(12,5))

sns.boxplot(
    x="hour",
    y="demand",
    data=train_df
)

plt.title("Hour of Day vs Demand")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    train_df.corr(
        numeric_only=True
    ),
    annot=True,
    cmap="Blues"
)

plt.title("Correlation Heatmap")

plt.show()
```

Exploratory Data Analysis revealed that RoadType and NumberofLanes are the most influential features affecting traffic demand. Highways and roads with 4–5 lanes exhibited substantially higher demand compared to residential roads and roads with fewer lanes. Temperature, day, and hour showed weak linear relationships with demand.
