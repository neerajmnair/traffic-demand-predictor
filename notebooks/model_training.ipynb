```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
train_df = pd.read_csv("../data/train.csv")
test_df = pd.read_csv("../data/test.csv")
```

Preprocessing

```python
roadtype_mode = train_df["RoadType"].mode()[0]

train_df["RoadType"] = train_df["RoadType"].fillna(
    roadtype_mode
)

test_df["RoadType"] = test_df["RoadType"].fillna(
    roadtype_mode
)
temp_mean = train_df["Temperature"].mean()

train_df["Temperature"] = train_df["Temperature"].fillna(
    temp_mean
)

test_df["Temperature"] = test_df["Temperature"].fillna(
    temp_mean
)
weather_mode = train_df["Weather"].mode()[0]

train_df["Weather"] = train_df["Weather"].fillna(
    weather_mode
)

test_df["Weather"] = test_df["Weather"].fillna(
    weather_mode
)

train_df["timestamp"] = pd.to_datetime(
    train_df["timestamp"],
    format="%H:%M"
)

test_df["timestamp"] = pd.to_datetime(
    test_df["timestamp"],
    format="%H:%M"
)

train_df["hour"] = train_df["timestamp"].dt.hour
train_df["minute"] = train_df["timestamp"].dt.minute

test_df["hour"] = test_df["timestamp"].dt.hour
test_df["minute"] = test_df["timestamp"].dt.minute

train_df.drop(
    "timestamp",
    axis=1,
    inplace=True
)

test_df.drop(
    "timestamp",
    axis=1,
    inplace=True
)
train_df.drop(
    "Index",
    axis=1,
    inplace=True
)

test_df.drop(
    "Index",
    axis=1,
    inplace=True
)
train_df = pd.get_dummies(
    train_df,
    drop_first=True
)

test_df = pd.get_dummies(
    test_df,
    drop_first=True
)

X_train = train_df.drop(
    "demand",
    axis=1
)

y_train = train_df["demand"]

X_train, test_df = X_train.align(
    test_df,
    join="left",
    axis=1,
    fill_value=0
)
```

```python
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train,
    y_train,
    test_size=0.2,
    random_state=42
)
```

```python
lr = LinearRegression()

lr.fit(X_tr, y_tr)

preds = lr.predict(X_val)

print("R2:", r2_score(y_val, preds))
print("MAE:", mean_absolute_error(y_val, preds))
```

```python
rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_tr, y_tr)

preds = rf.predict(X_val)

print("R2:", r2_score(y_val, preds))
print("MAE:", mean_absolute_error(y_val, preds))
```

```python
gb = GradientBoostingRegressor(
    random_state=42
)

gb.fit(X_tr, y_tr)

preds = gb.predict(X_val)

print("R2:", r2_score(y_val, preds))
print("MAE:", mean_absolute_error(y_val, preds))
```
